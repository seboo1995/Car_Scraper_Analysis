```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
base_url = "https://reklama5.mk/Search?city=&cat=24&q="
next_page = 'https://reklama5.mk/Search?city=&cat=24&q=&page=2'
num_of_pages = 10
```

```python
res = []
for page in range(1,num_of_pages):
    url = base_url + "&page="+str(page)
    request = requests.get(url)
    if request.status_code != 200:
        raise("Status code not 200")
    else:
        soup = BeautifulSoup(request.text)
        titles = soup.select('a.SearchAdTitle')
        links = ["https://reklama5.mk"+i["href"] for i in titles]
        naslov = [i.text for i in titles]
        for link in links:
            print("getting link: "+link)
            request_car = requests.get(link)
            if request_car.status_code != 200:
                print("Something happened with link: "+link)
                continue
            else:
                #### PRICE #######
                soup_car = BeautifulSoup(request_car.text)
                try:
                    price = soup_car.select("h5.mb-0")
                except:
                    try:
                        price = soup_car.select("div.col-md-7")
                    except:
                        price = 1
                #####################
                ##### OTHER INFO ########
                try:
                    other_info = soup_car.select(".card-body div.col-8")
                except:
                    print("The other info hit an exception"+link)
                    other_info = 'NA'
                    continue
                ###################
                #### Location #####
                try:
                    location = soup_car.select(".align-self-center p")
                except:
                    print("Location not found"+link)
                    location = "NA"
                    continue
                #### Text ####
                try:
                    text = soup_car.select("p.mt-3:nth-of-type(2)")
                except:
                    print("Text not found"+link)
                    text = "NA"
                    continue
                
            final = {
                'price':price,
                'other_info': other_info,
                'location': location,
                'text':text
            }
            print(final)
            res.append(final)
            time.sleep(3)
                    

                
                
                
                
        
```
